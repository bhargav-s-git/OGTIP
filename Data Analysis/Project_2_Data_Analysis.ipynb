```python
# Importing necessary libraries for analysis and visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the data
data = pd.read_csv('Data-Analysis_Oeson-data.csv')

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python
# Setting up visualization styles
sns.set(style="whitegrid")

# 1. Get the frequency of each year
year_frequency = data['work_year'].value_counts().sort_index()

# 2. Calculate average salaries for each year
average_salary_by_year = data.groupby('work_year')['salary_in_usd'].mean()

# Visualize the frequency of each year and average salaries for each year
fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()
ax1.bar(year_frequency.index.astype(str), year_frequency.values, color='g', alpha=0.6, label='Frequency')
ax2.plot(average_salary_by_year.index.astype(str), average_salary_by_year.values, color='b', marker='o', label='Average Salary')

ax1.set_xlabel('Year')
ax1.set_ylabel('Frequency', color='g')
ax2.set_ylabel('Average Salary in USD', color='b')

plt.title('Frequency of Data by Year and Average Salaries')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

The visualization shows the frequency of data entries for each year and the average salaries in USD for those years. From the bar graph (in green), we observe the distribution of data across different years, indicating the number of entries collected. The line graph (in blue) illustrates the average salary in USD for each year, providing insight into how the average salary trends over the years.

Next steps include:

 - Getting the frequency of job titles.
 - Identifying the top 10 most common data jobs.
 - Visualizing average salaries by job titles, highlighting those with the highest and lowest average salaries.

```python
# 3. Get the frequency of job titles
job_title_frequency = data['job_title'].value_counts()

# 4. Identify the top 10 most common data jobs
top_10_job_titles = job_title_frequency.head(10)

# Visualize the top 10 most common data jobs
plt.figure(figsize=(10, 8))
sns.barplot(x=top_10_job_titles.values, y=top_10_job_titles.index, palette="Blues_d")
plt.xlabel('Frequency')
plt.ylabel('Job Title')
plt.title('Top 10 Most Common Data Jobs')
plt.show()

top_10_job_titles
```

The visualization above displays the top 10 most common data jobs based on the dataset. The job titles are ranked by their frequency, with "Data Engineer," "Data Scientist," and "Data Analyst" being the most prevalent roles. This information is crucial for understanding the demand for specific data roles in the job market.

Next, we will:

 - Visualize average salaries by job titles, highlighting those with the highest and lowest average salaries.
 - Visualize average salaries for these most common job titles.

```python
# 5. Visualize average salaries by job titles
average_salary_by_job_title = data.groupby('job_title')['salary_in_usd'].mean().sort_values()

# Splitting for visualization clarity
lowest_salaries = average_salary_by_job_title.head(10)
highest_salaries = average_salary_by_job_title.tail(10)

# Visualizing the job titles with the highest and lowest average salaries
fig, axes = plt.subplots(2, 1, figsize=(10, 12))

sns.barplot(x=lowest_salaries.values, y=lowest_salaries.index, palette="Reds_d", ax=axes[0])
axes[0].set_title('Top 10 Job Titles with Lowest Average Salaries')
axes[0].set_xlabel('Average Salary in USD')
axes[0].set_ylabel('Job Title')

sns.barplot(x=highest_salaries.values, y=highest_salaries.index, palette="Greens_d", ax=axes[1])
axes[1].set_title('Top 10 Job Titles with Highest Average Salaries')
axes[1].set_xlabel('Average Salary in USD')
axes[1].set_ylabel('Job Title')

plt.tight_layout()
plt.show()
```

The visualizations above illustrate the job titles with the lowest and highest average salaries within the dataset. The first chart (top) shows the top 10 job titles with the lowest average salaries, indicating roles that, on average, pay less within the data domain. The second chart (bottom) displays the top 10 job titles with the highest average salaries, highlighting roles that are, on average, the most lucrative.

This differentiation is key for understanding the financial value associated with various data roles, providing insights into which job titles might be more financially rewarding and thus potentially more competitive or requiring higher levels of expertise.

Next, we will:

 - Visualize average salaries for the most common job titles identified earlier.

```python
# 6. Visualize Average Salaries for the most common job titles
# Filtering data to include only the top 10 most common job titles for this visualization
common_job_titles_data = data[data['job_title'].isin(top_10_job_titles.index)]
average_salary_for_common_jobs = common_job_titles_data.groupby('job_title')['salary_in_usd'].mean().sort_values()

plt.figure(figsize=(10, 8))
sns.barplot(x=average_salary_for_common_jobs.values, y=average_salary_for_common_jobs.index, palette="coolwarm")
plt.xlabel('Average Salary in USD')
plt.ylabel('Job Title')
plt.title('Average Salaries for the Most Common Job Titles')
plt.show()
```

The visualization presents the average salaries for the most common job titles, sorted by their average salary in USD. This analysis offers a clear perspective on how different roles within the data domain are valued in terms of compensation. It highlights not just the demand for certain positions (as seen in their frequency) but also how different roles are remunerated on average, providing valuable insights for both job seekers and recruiters in terms of salary expectations and budgeting for talent acquisition.

Moving forward, the subsequent analyses will include:

 - Visualizing average salaries based on employee residence and company location.
 - Exploring salary variations based on experience level.
 - Analyzing average salaries by employment type, work setting, and company size.

```python
# 10. Visualize average salary based on employee residence
average_salary_by_employee_residence = data.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(20)

# 10. Visualize average salary based on company location
average_salary_by_company_location = data.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(20)

fig, ax = plt.subplots(2, 1, figsize=(12, 16))

sns.barplot(x=average_salary_by_employee_residence.values, y=average_salary_by_employee_residence.index, palette="cubehelix", ax=ax[0])
ax[0].set_title('Top 20 Employee Residences by Average Salary in USD')
ax[0].set_xlabel('Average Salary in USD')
ax[0].set_ylabel('Employee Residence')

sns.barplot(x=average_salary_by_company_location.values, y=average_salary_by_company_location.index, palette="cubehelix", ax=ax[1])
ax[1].set_title('Top 20 Company Locations by Average Salary in USD')
ax[1].set_xlabel('Average Salary in USD')
ax[1].set_ylabel('Company Location')

plt.tight_layout()
plt.show()

```

The visualizations above showcase the top 20 employee residences and company locations by average salary in USD, providing insights into geographical variations in compensation for data-related jobs. The first chart (top) highlights the countries where employees reside that command the highest average salaries, indicating potentially higher cost of living or demand for data skills. The second chart (bottom) displays the top 20 company locations by average salary, reflecting where companies are based that offer the highest compensation, which might correlate with the company's revenue, industry, or local economic factors.

Continuing with our analysis, the next steps will involve:

 - Exploring salary variations based on experience level.
 - Analyzing average salaries by employment type, work setting, and company size.

```python
# 11. Visualize salary based on experience level
average_salary_by_experience = data.groupby('experience_level')['salary_in_usd'].mean().sort_values()

plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_by_experience.index, y=average_salary_by_experience.values, palette="magma")
plt.xlabel('Experience Level')
plt.ylabel('Average Salary in USD')
plt.title('Average Salary by Experience Level')
plt.show()
```

The bar chart illustrates the average salary by experience level, clearly showing how compensation tends to increase with higher levels of experience. This visualization underscores the value placed on experience within the data profession, with more senior roles typically commanding higher salaries. It highlights the financial growth potential within the field, providing a motivational insight for professionals at all stages of their careers to pursue skill development and experience accumulation.

Next, we will analyze and visualize:

 - Average salary based on employment type (e.g., Full-time, Part-time).
 - Average salary based on work setting (e.g., Remote, In-person, Hybrid).
 - Average salary by company size (e.g., Small, Medium, Large).

```python
# Adjusting data preparation
employment_type_series = pd.Series(average_salary_by_employment_type.index, name='Type')
work_setting_series = pd.Series(average_salary_by_work_setting.index, name='Type')
company_size_series = pd.Series(average_salary_by_company_size.index, name='Type')

average_salary_series = pd.concat([
    pd.Series(average_salary_by_employment_type.values, name='Average Salary in USD'),
    pd.Series(average_salary_by_work_setting.values, name='Average Salary in USD'),
    pd.Series(average_salary_by_company_size.values, name='Average Salary in USD')
], ignore_index=True)

category_series = pd.Series(['Employment Type']*len(average_salary_by_employment_type) +
                            ['Work Setting']*len(average_salary_by_work_setting) +
                            ['Company Size']*len(average_salary_by_company_size), name='Category')

type_series = pd.concat([employment_type_series, work_setting_series, company_size_series], ignore_index=True)

data_combined_corrected = pd.DataFrame({
    'Category': category_series,
    'Type': type_series,
    'Average Salary in USD': average_salary_series
})

# Visualizing the combined chart correctly
plt.figure(figsize=(12, 8))
sns.barplot(x='Average Salary in USD', y='Type', hue='Category', data=data_combined_corrected, palette='coolwarm')
plt.title('Combined Visualization: Average Salary by Employment Type, Work Setting, and Company Size')
plt.xlabel('Average Salary in USD')
plt.ylabel('Type / Size')
plt.legend(title='Category')
plt.show()
```

This visualization combines the data effectively, showcasing average salaries by employment type, work setting, and company size in a single chart. This approach enables a direct comparison across these categories, providing a comprehensive overview of how different factors influence salary levels within the data profession. Each category—employment type, work setting, and company size—is represented, offering valuable insights into salary dynamics based on these variables. This visualization serves as a useful tool for understanding the broader compensation landscape in data-related roles.
